```python
import psycopg2
import os

class DatabaseManager:
    def __init__(self, db_url):
        self.db_url = db_url

    def connect(self):
        """
        データベースに接続する。
        """
        try:
            self.connection = psycopg2.connect(self.db_url)
            self.cursor = self.connection.cursor()
            print("データベースに接続しました")
        except Exception as e:
            print(f"接続中にエラーが発生しました: {e}")

    def execute_sql_file(self, file_path):
        """
        SQLファイルを実行する。
        """
        try:
            with open(file_path, 'r') as file:
                sql_script = file.read()
            self.cursor.execute(sql_script)
            self.connection.commit()
            print(f"{file_path} のSQLが実行されました")
        except Exception as e:
            self.connection.rollback()
            print(f"SQL実行中にエラーが発生しました: {e}")

    def close(self):
        """
        データベース接続を閉じる。
        """
        if self.connection:
            self.cursor.close()
            self.connection.close()
            print("データベース接続を閉じました")
```

```python
db_url = "postgresql://{}:{}@postgres-service:{}/{}".format(os.environ['POSTGRES_USER'], os.environ['POSTGRES_PASSWORD'], os.environ['POSTGRES_SERVICE_SERVICE_PORT'], os.environ['POSTGRES_DB'])
db_manager = DatabaseManager(db_url)
db_manager.connect()
db_manager.execute_sql_file("create_table.sql")
db_manager.close()
```
